# **Propiedades de Logaritmo $\ln(x)$ y Exponencial  $\exp(x)$**

Estas funciones estan definidas en el modulo `numpy`.
Las propiedades matemáticas de estas funciones son las siguientes.


\begin{align*}
&Logaritmo&&Exponencial\\
\ln(a\cdot b)&=\ln(a)+ \ln(b)&  \exp(a+b) &=\exp(a)\cdot \exp(b)\\
\ln(\tfrac{a}{b})&=\ln(a)-\ln(b)&  \exp(a-b) &=\frac{\exp(a)}{\exp(b)}\\
\ln(a^b)&=b\cdot\ln(a)&  \big(\exp(a)\big)^b &=\exp(a\cdot b)\\
\ln\big(\exp(a)\big)&=a&  \exp\big(\ln(a)\big)&=a\\
\ln(1)&=0&  \exp(0) &=1\\
\end{align*}
Por otra parte, se tiene
$$\ln(x)\;\begin{cases}
			<0, & \text{si $0<x<1$},\\
            =0, & \text{si $x=1$},\\
            >0& \text{si $x>1$}.
		 \end{cases}$$
Análogamente

\begin{align*}
0<\exp(x)<1&&\text{si $x<0$},\\
\exp(x)=1&&\text{si $x=0$},\\
\exp(x)>1&&\text{si $x>0$}.
\end{align*}

## **Una aproximación de logaritmo importante en econometría**


Otra identidad que utilizaremos es la siguiente

$$\ln(x_1)-\ln(x_0)\approx\frac{x_1-x_0}{x_0}.$$

Antes de demostrar esta aproximación, recordemos que la definición de derivada de una función, esta la escribimos como

$$f'(x_0)=\lim_{x_1\to x_0}\frac{f(x_1)-f(x_0)}{x_1-x_0}.$$

Por lo tanto, para $f(x)=\ln(x)$ tenemos

<br>

\begin{align}
\frac{d\ln(x_0)}{dx}
&=\lim_{x_1\to x_0}\frac{\ln(x_1)-\ln(x_0)}{x_1-x_0}\\
&=\lim_{x_1\to x_0}\Big(\tfrac{1}{x_1-x_0}\Big)
\ln\Big(\tfrac{x_1}{x_0}\Big)\\
&=\lim_{x_1\to x_0}
\ln\Big(\tfrac{x_1}{x_0}\Big)^{\Big(\tfrac{1}{x_1-x_0}\Big)}.\\
\end{align}

Las identidades anterior, son únicamente propiedades de logaritmo. Enseguida, sumamos y restamos $x_0$ en el argumento de $\ln(\;)$ y multiplicamos por $\tfrac{x_0}{x_0}$ al exponente del argumento de logaritmo.
\begin{align}
\frac{d\ln(x_0)}{dx}
&=\lim_{x_1\to x_0}
\ln\Big(\tfrac{x_0+x_1-x_0}{x_0}\Big)
^{\Big(\tfrac{1}{x_1-x_0}\cdot\tfrac{x_0}{x_0}\Big)}\\
&=\lim_{x_1\to x_0}
\tfrac{1}{x_0}\cdot\ln\Big(1+\tfrac{x_1-x_0}{x_0}\Big)
^{\Big(\tfrac{x_0}{x_1-x_0}\Big)}\\
\end{align}

Denotamos $\alpha=\tfrac{x_1-x_0}{x_0}$ y observemos que cuando $x_1\to x_0$, sigue que $\alpha\to0$. Por otra parte, en cálculo diferencial es conocida la identidad de Euler
$$\lim_{\alpha\to 0}(1+\alpha)^{1/\alpha}=exp(1).$$

```python
from sympy import *

x = symbols('x')
funcion=(1+x)**(1/x);
print("Función : {}".format(funcion))

# Use sympy.limit() para calcular limites
Limit=limit(funcion, x,0)

print("El límite de Función cuando x tiende a 0 : {}".format(Limit))
```

Por lo anterior tenemos
$$\frac{d\ln(x_0)}{dx}
=\lim_{x_1\to x_0}
\tfrac{1}{x_0}\cdot\ln\Big(1+\tfrac{x_1-x_0}{x_0}\Big)
^{\Big(\tfrac{x_0}{x_1-x_0}\Big)}=\frac{1}{x_0}.$$

Finalmente, de la primer identidad de la definición de derivada de logaritmo tenemos
$$\frac{d\ln(x_0)}{dx}=\frac{1}{x_0}
=\lim_{x_1\to x_0}\frac{\ln(x_1)-\ln(x_0)}{x_1-x_0},$$
de la segunda identidad anterior se sigue que para "pequeñas" variaciones en $x$ se satisface la siguiente aproximación
$$\frac{x_1-x_0}{x_0}\approx\ln(x_1)-\ln(x_0).$$

El significado de "pequeño" es en función del contexto en que se trabaja, enseguida asignaremos un par de valores para ilustrar la precisión de la aproximación.

```python
# Ejemplo: elegimos los siguientes valores
import numpy as np

#x0=20
#x1=30

#x0=193
#x1=233

x0=40
x1=41

ratio=(x1-x0)/x0
DifLog=np.log(x1)-np.log(x0)

print("Cociente: {}".format(ratio))
print("Diferencia: {}".format(DifLog))
```

# **En econometría $\ln(x)$ representa una variación porcentual**

Continuando con esta aproximación
$$\frac{x_1-x_0}{x_0}\approx\ln(x_1)-\ln(x_0),$$
denotamos en términos generales

* $x_0=x,$
* $\Delta x=x_1-x_0,$
* $\ln(x+\Delta x)-\ln(x).$

En estos términos, la aproximación anterior queda como

$$\frac{\Delta x}{x}\approx\ln(x+\Delta x)-\ln(x),\qquad (\heartsuit)$$
esto significa que cuando el incremento $\Delta x$ es "pequeño" la diferencia de logaritmos representan cambios porcentuales. Por ejemplo, si
* $x=100,$
* $\Delta x=5,$

Tenemos
 * $$\frac{\Delta x}{x}=\frac{5}{100}=0.05$$
 * $$\ln(105)-\ln(100)=4.65-4.60=0.048$$

# **1. Modelo Log-lineal**


Comenzamos por considerar la forma funcional **log-lin**

$$\ln(y)=\beta_0+\beta_1 x.$$


<mark> En esta forma funcional el incremento $\Delta x$ corresponde a una  variación en $y$ de $(100\times \beta_1\Delta x )\%$. </mark>

<br>

En la forma funcional **log-lin** la variación en $x$ corresponde a una variación porcentual en $y$, esto lo deducimos de la aproximación de logaritmo $(\heartsuit)$ que demostramos previamente. En esta forma funcional dados  $x$ y el incremento $x+\Delta x$, los valores esperados son

$$\ln(y)=\beta_0+\beta_1x$$
y
$$\ln(y+\Delta y)=\beta_0+\beta_1 (x+\Delta x).$$

La diferencia de estos valores es
$$\ln(y+\Delta y)-\ln(y)=\beta_0+\beta_1 (x+\Delta x)-\beta_0-\beta_1 x
=\beta_1 \Delta x.$$

En la identidad anterior aplicamos la aproximación $(\heartsuit)$ y obtenemos
$$\frac{\Delta y}{y}=\beta_1 \Delta x.$$


La identidad anterior significa que el incremento $\Delta x$ es proporcional a una variación porcentual en $y$. Más precisamente, <mark>el incremento $\Delta x$ equivale en $y$ a la variación porcentual de $(100\times \beta_1\Delta x )\%$.</mark>

# **2. Modelo Lin-log**


El modelo **Lin-Log** es análogo al anterior, aplicamos logaritmo solo a la variable independiente $\ln(x)$ y dejamos lineal la variable dependiente. Es decir, dada la variable $x$ tenemos el valor de
$$y_1=\beta_0+ \beta_1 \ln(x),$$
ahora consideramos un incremento $\Delta x$ y denotamos el valor
$$y_2=\beta_0+ \beta_1 \ln(x+\Delta x).$$
De modo que

$$\Delta y=y_2-y_1
          =\Big(\beta_0+ \beta_1 \ln(x+\Delta x)\Big)-
           \Big(\beta_0+ \beta_1 \ln(x)\Big)
           =\beta_1 \Big(\ln(x+\Delta x)-\ln(x)\Big).
           $$

En la última identidad aplicamos la aproximación que hemos venido utilizando, de modo que tenemos

$$\Delta y=\beta_1 \frac{\Delta x}{x}.$$

<mark>
Esto significa que la variación porcentual $\Delta x/x$ corresponde con una variación en $y$ de $(\beta_1\times \frac{\Delta x}{100})$
</mark>

# **3. Modelo Log-log (elasticidad constante)**

Finalmente, para explicar la relación entre las variables, podemos aplicar logaritmo a la variable independiente $\ln(x)$ y a la variable dependiente $\ln(y)$, de modo que los estimadores los obtenemos de la regresión
$$\ln(y)=\beta_0+\beta_1\ln(x),$$
si consideramos el incremento $\Delta x$ en la variable independiente, tenemos el cambio en la variable $y$ es
$$\ln(y+\Delta y)=\beta_0+\beta_1\ln(x+\Delta x),$$
por lo tanto, la variación en $y$ queda como
\begin{align}
\ln(y+\Delta y)-\ln(y)
        &=\Big(\beta_0+\beta_1\ln(x+\Delta x)\Big)-\Big(\beta_0+\beta_1\ln(x)\Big)\\
        &=\beta_1\Big(\ln(x+\Delta x)-\ln(x)\Big).    
\end{align}
Aplicando la aproximación de logaritmo en la anterior ecuación tenemos
$$\frac{\Delta y}{y}=\beta_1\frac{\Delta x}{x}
\qquad\Rightarrow\qquad
\frac{\Delta y}{\Delta x}\cdot\frac{x}{y}=\beta_1.$$
Observemos que la última identidad del lado derecho es la definición de elasticidad de $y$ respecto a $x$, que mide la variación porcentual de $y$ sobre el cambio porcentual de $x$. La elasticidad es un concepto introducido en microeconomía para el análisis de las curvas de demanda; sin embargo, en muchas áreas de economía aplicada se plantean modelos de elasticidad constante, la regresión log-log nos premite especificar esos modelos. De modo que $\beta_1$ es la elasticidad de $y$ respecto a $x$, para $x,y>0$.

<mark>
La interpretación de la regresión log-log es la variación porcentual de  $\Delta x \%$ corresponde a una variación porcentual en $y$ de $(\beta_1\times\Delta x/100)\%$.
</mark>
